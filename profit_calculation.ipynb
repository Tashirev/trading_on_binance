```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
start_volume = 90000
day_profit_percent = 0.2
day_of_work = 30*12
print(start_volume)
```

```python
trade = pd.DataFrame({'day': 0, 'volume': start_volume},index=[0])

volume = start_volume
for i in range(1,day_of_work+1):
    volume = volume + volume*day_profit_percent/100
    trade = pd.concat([trade, pd.DataFrame({'day': i, 'volume': volume},index=[i])], ignore_index=True)

trade.iloc[-1]
```

```python
sns. lineplot(data=trade, x='day', y='volume')
```

## Подбор коэффициента шага при сделках

```python
# n - число предыдущих сделок
n = np.array(range(20))
# m - коэффициент
m = 1.15**n
sns.lineplot(x=n,y=m)
```

```python
v = 300
part = 0.04
v_list = list()
v_trade_list = list()
for i in range(20):
    i = i+1
    v_trade = v*part*i
    v_list.append(v)
    v = v - v_trade
    v_trade_list.append(v_trade)
v = np.array(v_list)
v_trade = np.array(v_trade_list)
sns.lineplot(x=n,y=v)
sns.lineplot(x=n,y=v_trade)
print(v_trade)
```
